```python
import pandas as pd
# Load the tornado dataset
tornado = pd.read_csv('/content/drive/MyDrive/Fall 2025/Colab Notebooks/tornados.csv')
tornado.head()
```

```python
# Display the shape of the dataframe (number of rows and columns)
tornado.shape
```

```python
# Display the data for the 5th row (index 4)
tornado.iloc[4]
```

```python
# Count the occurrences of each tornado magnitude in Illinois
tornado[tornado.st=="IL"].mag.value_counts() #Illinois
```

```python
# Count the occurrences of each tornado magnitude in Florida
tornado[tornado.st=="FL"].mag.value_counts() #Florida
```

```python
# Count the occurrences of each tornado magnitude in Oklahoma
tornado[tornado.st=="OK"].mag.value_counts() #Oklahoma
```

```python
# Count the occurrences of the strongest tornadoes (magnitude 5) by state
tornado[tornado.mag==5].st.value_counts()
```

```python
# Find the state with the most injuries from a single tornado
tornado.iloc[tornado.inj.argmax()].st
```

```python
# Find the state with the most fatalities from a single tornado
tornado.iloc[tornado.fat.argmax()].st
```

```python
# Find the state with the highest property loss from a single tornado
tornado.iloc[tornado.loss.argmax()].st
```

```python
# Count the states that appear in the top 5 tornadoes with the most injuries
tornado.iloc[tornado.inj.nlargest(5).index].st.value_counts()
```

```python
# Find the deadliest tornado in Illinois
tornado[tornado.st=="IL"].iloc[tornado[tornado.st=="IL"].inj.argmax()]
```
