```python
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris()
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(iris.data, iris.target).\
predict(iris.data)
datasize=iris.data.shape[0]
mislabel = (iris.target != y_pred).sum()
print('Number of mislabeled points out of a total %d points : %d' %(datasize,mislabel))
print(pd.crosstab(gnb.predict(iris.data), iris.target))
```
